```python
path = r'C:\Users\User\Desktop\j\lessons\lesson_4\Res_Tree\F000545\res_2019.09.10_0.0_6493B6_Container-dat_2110_84-99-223_F000545\meals_list_2.txt'
```

```python
data = open(path).readlines()
data
```

```python
all_data = []
scanning = False
for i in data:
    if i.startswith('-----------------------'):
        scanning = False
        continue
        
    if scanning:
        if 'Meal#' in i:
            columns_name = i
            continue
        all_data.append([i.strip() for i in i.split(';')])
        
    if 'Non-deleted saved meals' in i:
        scanning = True
        continue
all_data
```

```python
columns_name_final = [col.strip() for col in columns_name.split(';')]
columns_name_final
```

```python
df = pd.DataFrame(all_data, columns = columns_name_final[0:17])
df.head()
```

```python
df['Date'] = pd.to_datetime(df.Date, dayfirst = True)
```

```python
df.dtypes
```

```python
df = df.astype({'Minutes':'int', 'Pro':'float'})
user_df = df[['Date', 'Minutes', 'Pro']]
user_df.head()
```

```python
user_df.groupby('Date', as_index = False) \
    .agg({'Minutes':'sum'})
```
