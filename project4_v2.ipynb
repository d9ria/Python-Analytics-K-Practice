```python
from pathlib import Path
import pandas as pd
import os
```

```python
path = Path(r'C:\Users\User\Desktop\j\lessons\lesson_4\data')
path
```

```python
df = pd.DataFrame()
for current_path, dirs, files in os.walk(path):
    for file in files:
        data_path = Path(current_path) / file
        temp_df = pd.read_csv(data_path)
        temp_df['name'] = data_path.name
        temp_df['date']= data_path.parts[-2]
        df = pd.concat((df, temp_df))
df
```
